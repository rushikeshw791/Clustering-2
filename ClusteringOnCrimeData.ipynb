**Hierarchy Clustering**

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
crime = pd.read_csv('https://github.com/rushikeshw791/Clustering-2/files/9819012/crime_data.csv')
crime
```

```python
# normalize the data function
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
# creating clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
crime['h_clusterid']=hc.labels_
crime
```

**KMeans clustering**

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1= pd.read_csv('https://github.com/rushikeshw791/Clustering-2/files/9819012/crime_data.csv')
```

```python
crime1
```

```python
# Normalized data function
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(crime1.iloc[:,1:])
```

```python
# Elbow Curve
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df_norm)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2= pd.read_csv('https://github.com/rushikeshw791/Clustering-2/files/9819012/crime_data.csv')
```

```python
crime2.info()
```

```python
df=crime2.iloc[:,1:5]
```

```python
df.values
```

```python
stdscaler=StandardScaler().fit(df.values)
x=stdscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
c1
```

```python
pd.concat([crime2,c1],axis=1)
```
